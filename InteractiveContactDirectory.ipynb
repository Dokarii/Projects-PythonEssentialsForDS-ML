```python
contacts = {'Richard': '30230230', 'Ann': '30130130', 'Louis': '30030030'}
print('1. Add a contact\n2. Search for a contact\n3. Edit a contact\n4. Delete a contact\n5. Show all contacts')
```

```python
option = int(input('Which option do you prefer (1, 2, 3, etc.): '))
if option == 1:
    name, number = input('Contact name: '), input('Contact number: ')
    contacts[name] = number
    print(contacts)
elif option == 2:
    name = input('Contact name: ')
    if name in contacts:
        print(contacts[name])
    else:
        print('Error! User not found')
elif option == 3:
    name, number = input('Name of the contact you want to edit: '), input('New contact number: ')
    if name in contacts:
        contacts[name] = number
        print(f'Number updated: {contacts[name]}')
    else:
        print('Error! The contact you entered does not exist')
elif option == 4:
    name = input('Name of the contact to delete: ')
    if name in contacts:
        del contacts[name]
        print('Contact successfully deleted')
    else:
        print('Error! The contact you entered does not exist or has already been deleted')
else:
    print(contacts)
```
