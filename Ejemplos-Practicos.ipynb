```python
info_ciudades = (("Buenos Aires", 3000000, "Argentina"),("Madrid", 3200000, "España"),("Tokio", 13929286, "Japón"))

# Ciudad 1: Buenos Aires
nombre_0 = info_ciudades[0][0]
poblacion_0 = info_ciudades[0][1]
pais_0 = info_ciudades[0][2]
print(f"La ciudad de {nombre_0} en {pais_0} tiene una población de {poblacion_0} habitantes")

# Ciudad 2: Madrid
nombre_1 = info_ciudades[1][0]
poblacion_1 = info_ciudades[1][1]
pais_1 = info_ciudades[1][2]
print(f"La ciudad de {nombre_1} en {pais_1} tiene una población de {poblacion_1} habitantes")

# Ciudad 3: Tokio
nombre_2 = info_ciudades[2][0]
poblacion_2 = info_ciudades[2][1]
pais_2 = info_ciudades[2][2]
print(f"La ciudad de {nombre_2} en {pais_2} tiene una población de {poblacion_2} habitantes")
```

```python
dic = {"Ana" : 30, "Luis" : 25, "María" : 28}

dic["Ana"] = 31
print(dic)

dic.update({"Luis": 26})
print(dic)
```

```python
dic = {"Ana": 30, "Luis": 25}

nuevos_datos = [("María", 28), ("Pedro", 35)]

dic.update(nuevos_datos)
print(dic) 
# Resultado: {'Ana': 30, 'Luis': 25, 'María': 28, 'Pedro': 35}
```

```python
dic = {"Ana": 30, "Luis": 25}

dic.update(Maria=28, Pedro=35, Luis=26)

print(dic)
# Resultado: {'Ana': 30, 'Luis': 26, 'Maria': 28, 'Pedro': 35}
```

```python
a = 10
b = 5
c = 9

print(a > b and a > c)  # True
print(a > b or a > c)   # True
if not(a < b or a < c):  # False
    print("No se cumple la condición")
print(a != b)  # True
```
