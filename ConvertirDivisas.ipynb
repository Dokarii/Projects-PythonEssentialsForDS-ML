```python
from datetime import datetime
name = input("Enter your name: ")
date = datetime.now()
```

```python
amount_usd = float(input("Enter the amount in USD to convert: "))
```

```python
exchange_rate = 0.85
euros_to_receive = exchange_rate * amount_usd
```

```python
bills_10 = int(euros_to_receive // 10)
remaining_euros = euros_to_receive % 10
```

```python
bills_1 = int(remaining_euros // 1)
remaining_euros = remaining_euros % 1
```

```python
coins = int(remaining_euros // 0.01)
```

```python
print(f"Hello {name}. Welcome to the currency converter. Today is {date.strftime('%Y-%m-%d')}")
print(f"Amount to receive: {euros_to_receive:.2f} EUR")
print(f"10 EUR bills: {bills_10}")
print(f"1 EUR bills: {bills_1}")
print(f"1 cent coins: {coins}")
```
