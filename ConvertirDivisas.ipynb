```python
from datetime import datetime
nombre = input("Ingrese su nombre: ")
fecha = datetime.now()
```

```python
cantidad_usd = float(input("Ingrese la cantidad en USD que desea convertir: "))
```

```python
tasa_cambio = 0.85
euros_a_recibir = tasa_cambio * cantidad_usd
```

```python
billetes_de_10 = int(euros_a_recibir // 10)
euros_restantes = euros_a_recibir % 10
```

```python
billetes_1 = int(euros_restantes // 1)
euros_restantes = euros_restantes % 1
```

```python
monedas = int(euros_restantes // 0.01)
```

```python
print(f"Hola {nombre}. Bienvenido al convertidor de divisas. Hoy es {fecha.strftime('%Y-%m-%d')}")
print(f"Cantidad a recibir: {euros_a_recibir:.2f} EUR")
print(f"Billetes de 10 EUR: {billetes_de_10}")
print(f"Billetes de 1 EUR: {billetes_1}")
print(f"Monedas de 1 centavo: {monedas}")
```
